```python
import gradio as gr
import ollama

# Function to interact with Phi-3
def chat_with_phi3(message, history):
    # Define a system prompt to make it a medical chatbot
    system_prompt = "You are a helpful medical chatbot. Provide accurate and simple answers to health-related questions. If unsure, suggest consulting a doctor."

    # Prepare the input for Ollama
    response = ollama.chat(
        model="phi3",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": message}
        ]
    )

    # Return the response
    return response['message']['content']

# Create the Gradio interface
interface = gr.ChatInterface(
    fn=chat_with_phi3,
    title="AI Medical Chatbot",
    description="Ask me anything about health! Note: This is not a substitute for professional medical advice.",
    theme="soft"
)

# Launch the interface
interface.launch()
```

```python
import gradio as gr
import ollama
from pyngrok import ngrok

# Function to interact with Phi-3
def chat_with_phi3(message, history):
    system_prompt = "You are a helpful medical chatbot. Provide accurate and simple answers to health-related questions. If unsure, suggest consulting a doctor."
    response = ollama.chat(
        model="phi3",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": message}
        ]
    )
    return response['message']['content']

# Create the Gradio interface
interface = gr.ChatInterface(
    fn=chat_with_phi3,
    title="AI Medical Chatbot",
    description="Ask me anything about health! Note: This is not a substitute for professional medical advice.",
    theme="soft"
)

# Launch with Ngrok
public_url = ngrok.connect(7860)  # Gradio’s default port
print(f"Public URL: {public_url.public_url}")
interface.launch(server_name="0.0.0.0", server_port=7860)
```
